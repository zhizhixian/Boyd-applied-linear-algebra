```python
import numpy as np
import cv2

def image_new(image,n):  #n为保留的奇异值个数
    u,s,v=np.linalg.svd(image)  #u=m*m  left matrix;v=n*n right matrix
    s1=np.zeros(image.shape)
    for i in range(n):
        s1[i][i]=s[i]  #s1 diagonal matrix
    img=u@s1@v
    img=img.astype(np.uint8) #opencv-python中为图像创建一个容器，需指定dtype=np.uint8
    return img

if  __name__ == "__main__":
    bul=cv2.imread('building.jpeg') 
    print(bul.shape)
    #直接 bul_new=image_new(bul,10)  显示错误
    
    a=bul[: , : ,0]
    b=bul[: , : ,1]
    c=bul[: , : ,2]
    a1=image_new(a,400)
    b1=image_new(b,400)
    c1=image_new(c,400)
    bul_new = np.stack([a1,b1,c1], axis=2)
    print(bul_new.shape)
    
    cv2.imwrite("./bul_new400.jpeg", bul_new)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
