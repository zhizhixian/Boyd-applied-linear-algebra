```python
#exercise 9.2: dynamics of an economy
import numpy as np
T=20
B=np.array([[0.1,0.06,0.05,0.7],[0.48,0.44,0.1,0.04],[0,0.55,0.52,0.04],[0.04,0.01,0.42,0.51]])
x_1=np.array([0.6,0.9,1.3,0.5])
state_traj=np.column_stack([x_1,np.zeros((4,T-1))])
total=np.zeros((T))
for t in range(T-1):
    state_traj[:,t+1]=B@state_traj[:,t]
for i in range(T):
    total[i]=sum(state_traj[:,i])
#plot the dynamics    
import matplotlib.pyplot as plt
plt.ion()
plt.plot(np.arange(T),state_traj.T)
plt.legend(['(x_t)_1','(x_t)_2','(x_t)-3','(x_t)_4'])
plt.xlabel('year')
plt.ylabel('output')
plt.show()
#plot the total output versus t
plt.plot(np.arange(T),total.T)
plt.xlabel('year')
plt.ylabel('total output')
plt.show()
```

```python
import numpy as np
T=10
A1=np.array([[0.1,0.05,0.2],[0.2,0.01,-0.21],[0.5,-0.12,0.31]])
A2=np.array([[-0.11,0.6,0.4],[0.2,-0.4,0.14],[0.7,-0.3,0.22]])
x_1=np.array([1,1,1])
x_2=np.array([0.5,1,0.5])
state_trajx=np.column_stack([x_1,x_2,np.zeros((3,T-2))])
for t in range(T-2):
    state_trajx[:,t+2]=A1@state_trajx[:,t+1]+A2@state_trajx[:,t]
    
B=np.block([
    [A1       ,A2             ],
    [np.eye(3),np.zeros((3,3))]
])
z_1=np.block([x_2,x_1]) #注意z_1和z_2的上下位置关系！！！！！！
state_trajz=np.column_stack([z_1,np.zeros((6,T-1))])
for t in range(T-1):
    state_trajz[:,t+1]=B@state_trajz[:,t]
    
import matplotlib.pyplot as plt
plt.ion()
plt.plot(np.arange(T),state_trajx.T)
plt.legend(['x_1','x_2','x_3'])
plt.show()

plt.plot(np.arange(T),state_trajz.T)
plt.show()
#两图中绿色、蓝色、黄色三条线完全重合！！！！
```
